```python
df["Income"] = pd.to_numeric(df["Income"].str.replace(",",""))
```

```python
#df = df.drop_duplicates()
df = df.drop_duplicates(subset=["ID"])
```

```python
columnas = ["Marital Status", "Gender", "Education", "Occupation", "Commute Distance", "Region"]
for col in columnas:
    df[col] = df[col].str.strip().str.replace("Miles","").str.lower()
```

```python
df = df.dropna(subset=["Age"])
```
